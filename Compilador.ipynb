```python
!# Função para saber em qual aresta o caracter deve prosseguir

def pertence(string):
   if string in alfabeto:
        if string in D: return 'D'
        elif string in L:
            if string in Exp: return 'Exp'
            else: return 'L'
        elif string in A2: return 'A2'
        else:
            return string
   else: return 'nulo'
```

```python
def call_Token(estado, string):
    if estado == 1 or estado == 3 or estado == 6:
        return 'num'
    elif estado == 7:
        if (string) in tabela_simbolos:
          if tabela_simbolos[string] == ('id','Nulo'):
            return 'id'
          else: return string
        else: return 'id'
    elif estado == 9:
        return 'lit'
    elif estado == 10 or estado == 11 or estado == 12:
        return 'opr'
    elif estado == 13:
        return 'opm'
    elif estado == 14:
        return 'rcb'
    elif estado == 16:
        return 'comentário'
    elif estado == 17:
        return 'fc_p'
    elif estado == 18:
        return 'ab_p'
    elif estado == 19:
        return 'pt_v'
    elif estado == 20:
        return 'vir'
    elif estado == 21:
        return 'EOF'
    else :
        return 'erro'
```

```python
def dfa(caracter, atual, arestas):
  c = pertence(caracter)
  if (atual, c) in arestas:
   next_states = arestas[(atual, c)]
   atual = next_states[0]
   return atual
  else:
    return "Nulo"
```

```python
def analise_lex(caracteres_lidos, arestas, atual):
    # entao os caracteres_lidos é uma lista de car[caractere,linha,coluna]
    i=0             #iterador while para percorrer a lista de car
    l=[]
    palavra=[]
    al=[]
    while i < len(caracteres_lidos):
      # joga na função dfa o caractere de cada car a medida que o i incrementa
      atual=dfa(caracteres_lidos[i][0],atual,arestas)
      #atual = estado do caractere no dfa
      # se não for literal ou comentário:
      if not (atual==8 or atual==15):

        # se for '\n' , ' ' ou '\t' , passa para o caractere seguinte
        if caracteres_lidos[i][0] == '\n'or caracteres_lidos[i][0] == ' ' or caracteres_lidos[i][0] == '\t':
          i+=1
          continue

        # se não for '\n' , ' ' ou '\t'
        else:
          l.append(caracteres_lidos[i][0]) # adiciona o caractere no l
          # se não for o último caractere do arquivo:
          if i+1 < len(caracteres_lidos):
            c=pertence(caracteres_lidos[i+1][0])
            #se não existe um próximo estado com esse caractere
            if not ((atual, c) in arestas):
              palavra.append(''.join(l))
              l=[]
              car = [palavra, caracteres_lidos[i][1], caracteres_lidos[i][2], atual]
              atual=0
              al.append(car)
              palavra=[]
          # se for o último caractere do arquivo:
          else:
            palavra.append(''.join(l))
            l=[]
            car = [palavra, caracteres_lidos[i][1], caracteres_lidos[i][2], atual]
            atual=0
            al.append(car)
            palavra=[]

     # se for literal ou comentário:
      else:
        l.append(caracteres_lidos[i][0])  # adiciona o caractere no l
        if i+1 < len(caracteres_lidos):
          c=pertence(caracteres_lidos[i+1][0])
          if not (atual, c) in arestas:
            palavra.append(''.join(l))
            l=[]
            car = [palavra, caracteres_lidos[i][1], caracteres_lidos[i][2], atual]
            atual=0
            al.append(car)
            palavra=[]
        else:
          palavra.append(''.join(l))
          l=[]
          car = [palavra, caracteres_lidos[i][1], caracteres_lidos[i][2], atual]
          atual=0
          al.append(car)
          palavra=[]

      i+=1 #incrementa o while
    return al
```

```python
def tipo_tupla(estado,string):
    if estado == 1: # inteiro do num
        return 'inteiro'
    elif estado == 3 or estado == 6: # real do num
        return 'real'
    elif estado == 9:
        return 'literal'
    elif estado == 7:
        if (string) in tabela_simbolos:
          if tabela_simbolos[string] == ('id','Nulo'):
            return 'Nulo'
          else: return string
        else:
            tabela_simbolos[string] = ('id','Nulo') # adiciona na tabela de símbolos a id , string , nulo
            return 'Nulo'
    else:
        return 'Nulo'
```

```python
def recuperação_erro(token_anterior, i):
    # Recuperação de erro em modo pânico

     while True:
          if i >= (len(tok) - 1) or tok[i][0] in ['literal','real','inteiro','repita','se','$']:
              break
          i += 1

          if tok[i][0] != token_anterior:
              a = tok[i][0]
              break

      # Se todos os tokens restantes são iguais ao token anterior, avançamos para o próximo token
     if i < (len(tok) - 1) and tok[i][0] == token_anterior:
          i += 1
          a = tok[i][0]

     return a, i
```

```python
def operadores(op1, op, op2):
  if op == '+':
    return op1+op2
  elif op == '-':
    return op1-op2
  elif op == '*':
    return op1*op2
  elif op == '/':
    return op1/op2
  elif op == '<':
    if op1<op2:
      return True
    else: return False
  elif op == '>':
    if op1>op2:
      return True
    else: return False
  elif op == '<=':
    if op1<=op2:
      return True
    else: return False
  elif op == '>=':
    if op1>=op2:
      return True
    else: return False
  elif op == '=':
    if op1==op2:
      return True
    else: return False
```

```python
def analise_SRL():
  semantico = []
  i = 0
  a = tok[i][0]
  while True:
    if tok[i][0] == 'erro' or tok[i][0] == 'comentário':
      if tok[i][0] == 'erro':
        E_L(tok[i])
        er[0]+=1
      del tok[i]
      a = tok[i][0]
      continue;
    # Seja s o estado no topo da pilha
    s = pilha[-1]

    R = (AC.loc[s,a]).split('R')
    S = (AC.loc[s,a]).split('S')
    #print("a: ",a)
    #print("s: ",s)
    #print("Action: ", AC.loc[s,a])
    #print("t: ",AC.loc[s,a][1])

    # Verifique ação de deslocamento (shift)
    if S[0] == '':
      t = int(S[1])
      # Empilha t na pilha
      pilha.append(t)

      if i<(len(tok)-1):
        #print("----->",i,len(tok))
        i += 1
        a = tok[i][0]
      if (a != 'inicio' and a != 'varinicio' and a != 'varfim' and a != '$' and a != 'num' and a != 'erro'):
        semantico.insert(0, tok[i][1])
      if a == 'num':
        if tok[i][2] == 'inteiro':
          semantico.insert(0, (int(tok[i][1]), tok[i][2]))
        if tok[i][2] == 'real':
          semantico.insert(0, (float(tok[i][1]), tok[i][2]))

    # Verifique ação de redução (reduce)
    elif R[0] == '':
      # reduce A->B
      B = Prd[int(R[1])].split('→')[1]
      card_B = len(B.split(' '))
      #print("B: ",card_B)
      # Desempilha símbolos |B| da pilha
      for j in range(card_B):
        pilha.pop()

      # Empilhe GOTO[t,A] na pilha
      A = Prd[int(R[1])].split('→')[0]
      t = pilha[-1]
      #print("GOTO: ",int(GT.loc[t, A]))
      pilha.append(int(GT.loc[t, A]))
      #print("A: ",A)
      print(Prd[int(R[1])])
      if i<(len(tok)-1): Sem(Prd[int(R[1])], semantico, tok[i][3])

    # Verifique ação de aceitação
    elif AC.loc[s,a] == 'Acc':
      print("Programa finalizado com sucesso!")
      # Pare - a análise terminou
      break
    elif AC.loc[s,a] == 'erro':
      er[0]+=1
      if i >= len(tok)-1:
        print("Erro: Programa finalizou e não entrou no estado de Aceite")
        break
      if s == 0 and a!= 'inicio':
        print("Não foi identificada a palavra reservada inicio")
        objeto = ['inicio', 'inicio', 'inicio']
        tok.insert(0, objeto)
        a = tok[i][0]
      elif s == 2 and a!= 'varinicio':
        print("Não foi identificada a palavra reservada varinicio")
        objeto = ['varinicio', 'varinicio', 'varinicio']
        tok.insert(1, objeto)
        a = tok[i][0]
      else:
        print(f"Erro: lexema {a} na linha {tok[i][3]}, coluna {tok[i][4]}")
        token_anterior = a
        a, i = recuperação_erro(token_anterior, i)
```
